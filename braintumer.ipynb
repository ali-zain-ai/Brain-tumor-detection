```python
import pandas as pd
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.applications.vgg16 import VGG16
from tensorflow import keras
import tensorflow as tf
```

```python
path_di = r"D:\deep learning\brain_tumor_dataset"

data = []
labels = []

for category in ['yes', 'no']:
    path_foler = os.path.join(path_di, category)
    label = 1 if category == 'yes' else 0 
    

    for filename in os.listdir(path_foler):
            image_path = os.path.join(path_foler, filename)
            # print(image_path)
            # break
            img = cv2.imread(image_path)
            img = cv2.resize(img, (224, 224))
            data.append(img)
            labels.append(label)

# print(image_path[1], labels[1])
```

```python
X = np.array(data)
# print(X[1])
y = np.array(labels)
print(y[1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
print(X_train[10], y_train[10])
```

```python
vgg = VGG16()
print(vgg.summary())
```

```python
from keras import Sequential
model = Sequential()

for layer in vgg.layers[:-1]:
    model.add(layer)

# print(model.summary())

for layer in model.layers:
    layer.trainable = False

from keras.layers import Dense
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

```python
model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=6, validation_data=(X_test, y_test))
```

```python
def brain_tumor(sample_image):
    y_pred = model.predict(sample_image.reshape(1, 224, 224, 3))
    return y_pred[0][0]
```

```python
sample1 = cv2.imread("D:\deep learning\Y40.JPG")
sample2 = cv2.resize(sample1, (224, 224))
brain_tumor(sample2)
```
